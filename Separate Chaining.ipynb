```python
class Person815: 
    def __init__(self):
        self.nic = ""
        self.name = ""
        self.bday = ""
        self.addrss = ""
        # creating object to reference SepChain class
        self.sc = SepChain815()        
        
    def setObject(self, nic, name, bday, addrss):
        self.nic = nic
        self.name = name
        self.bday = bday
        self.addrss = addrss
        #creating person object as a tuple        
        obj = (self.nic, self.name, self.bday, self.addrss)        
        return self.sc.insert(obj)
        
    def getObject(self, nic):
        self.nic = nic
        return self.sc.search(self.nic)
    
    def deleteObject(self, nic):
        self.nic = nic
        return self.sc.delete(self.nic)
```

```python
class SepChain815:
    
    def __init__(self):
        # creating list
        self.hashList = [None]*101
        
    # hashing function    
    def hash(self,nic):
        # nic is a non random key. breaking into pairs and adding them together
        nicList = [i for i in nic]
        sum = int(nicList[0] + nicList[1]) + int(nicList[2] + nicList[3]) + int(nicList[4] + nicList[5]) + int(nicList[6] + nicList[7]) + int(nicList[8])
        #list size is 101
        key = sum % 101  
        return key
    
    # insert function
    def insert(self, obj):        
        # calling hash function
        key = self.hash(obj[0])        
        
        # checking for key availability and adding or appending the new object
        if self.hashList[key] == None:
            self.hashList[key] = [obj]
        else:
            self.hashList[key].append(obj)
        return self.hashList
    
    # search function
    def search(self, nic):
        # calling hash function
        key = self.hash(nic)
        
        # checking for key availability and presenting result set or returning False if not found
        if self.hashList[key] == None:            
            return False           
        else:
            for i in self.hashList[key]:
                if(nic == i[0]):                    
                    return [i[1], i[2], i[3]]
                
    # delete function
    def delete(self, nic):
        # calling hash function
        key = self.hash(nic)
        
        # checking for key availability and deletion or returning False if not found
        if self.hashList[key] == None:            
            return False           
        else:
            for i in self.hashList[key]:
                if(nic == i[0]):                     
                    self.hashList[key].remove(i)
                    print("Deleted")
                    return self.hashList
```
