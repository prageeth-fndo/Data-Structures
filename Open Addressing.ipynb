```python
class OpenAddrss815:
    
    def __init__(self):
        self.listSize = 101
        self.hashList = [None] * self.listSize 
        
    # hashing function    
    def hash(self,nic):
        # nic is a non random key. breaking into pairs and adding them together
        nicList = [i for i in nic]
        sum = int(nicList[0] + nicList[1]) + int(nicList[2] + nicList[3]) + int(nicList[4] + nicList[5]) + int(nicList[6] + nicList[7]) + int(nicList[8])
        #default list size is 101
        key = sum % self.listSize
        return key    
    
    # linear probing
    ####################################################################################################################
    
    #array size exceeding function
    def linearExceeding(self):        
        tempList = []
        #checking for array size
        for x in self.hashList:
            #counting objects and inserting them into a temporary list
            if x != None:
                tempList.append(x)
        if(len(tempList) >= len(self.hashList)/3*2):
                self.hashList = [None] * len(self.hashList) * 2
                self.listSize = len(self.hashList)                
                #copying objects from temporary list to original hash list
                for item in tempList:
                    self.insertLinear(item)        
    
    # insert function
    def insertLinear(self, obj):               
        # calling hash function
        key = self.hash(obj[0])
        # checking for key availability and adding the new object       
        while(key < len(self.hashList)):
            if self.hashList[key] == None:
                self.hashList[key] = obj
                #calling linearExceeding function
                self.linearExceeding()
                return self.hashList
            else:                           
                # incrementing the key value
                key = key + 1
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList)
        
    # search function
    def searchLinear(self, nic):
        # calling hash function
        key = self.hash(nic)
        # checking for key availability and presenting result set or returning False if not available in the list
        while(key < len(self.hashList)):
            if self.hashList[key] == None:
                return False            
            if(nic == self.hashList[key][0]):
                return [self.hashList[key][1], self.hashList[key][2], self.hashList[key][3]]
            else:
                # incrementing the key value
                key = key + 1
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList)
        else:
            return False        
    
    # delete function
    def deleteLinear(self, nic):
        # calling hash function
        key = self.hash(nic)        
        # checking for key availability and deletion or returning False if not found
        while(key < len(self.hashList)):
            if self.hashList[key] == None:
                return False            
            if(nic == self.hashList[key][0]):
                # making the index to None
                self.hashList[key] = None
                print("Deleted")
                return self.hashList
            else:
                # incrementing the key value
                key = key + 1
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList)
        else:
            return False
    
    # end of linear probing
    ####################################################################################################################
      
    
    
    # quadratic probing
    ####################################################################################################################
    
    #array size exceeding function
    def quadExceeding(self):        
        tempList = []
        #checking for array size
        for x in self.hashList:
            #counting objects and inserting them into a temporary list
            if x != None:
                tempList.append(x)
        if(len(tempList) >= len(self.hashList)/3*2):
                self.hashList = [None] * len(self.hashList) * 2
                self.listSize = len(self.hashList)
                
                #copying objects from temporary list to original hash list
                for item in tempList:
                    self.insertQuad(item)
                
    # insert function
    def insertQuad(self, obj):        
        # calling hash function
        key = self.hash(obj[0])        
        # checking for key availability and adding the new object
        x = 1; y = key
        while(key < len(self.hashList)):            
            if self.hashList[key] == None:
                self.hashList[key] = obj
                #calling linearExceeding function
                self.quadExceeding()
                return self.hashList
            else:
                # increment key value by sequence of 1, 4, 9, 16.....                
                key = y + (x)**2
                x = x + 1                
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList) 
        else:
            return False
                    
    # search function
    def searchQuad(self, nic):
        # calling hash function
        key = self.hash(nic)       
        # checking for key availability and presenting result set or returning False if not available in the list
        x = 1; y = key
        while(key < len(self.hashList)):
            if self.hashList[key] == None:
                return False            
            if(nic == self.hashList[key][0]):                    
                return [self.hashList[key][1], self.hashList[key][2], self.hashList[key][3]] 
            else:
                # increment key value by sequence of 1, 4, 9, 16.....                
                key = y + (x)**2
                x = x + 1
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList)
        else:
            return False    
    
    # delete function
    def deleteQuad(self, nic):
        # calling hash function
        key = self.hash(nic)        
        # checking for key availability and deletion or returning False if not found
        x = 1; y = key
        while(key < len(self.hashList)):
            if self.hashList[key] == None:
                return False            
            if(nic == self.hashList[key][0]):
                # making the index to None
                self.hashList[key] = None
                print("Deleted")
                return self.hashList
            else:
                # increment key value by sequence of 1, 4, 9, 16.....
                key = y + (x)**2
                x = x + 1
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList)
        else:
            return False
    
    # end of quadratic probing
    ####################################################################################################################
    
    
    
    # double hashing
    ####################################################################################################################
    
    #array size exceeding function
    def doubleExceeding(self):        
        tempList = []
        #checking for array size
        for x in self.hashList:
            #counting objects and inserting them into a temporary list
            if x != None:
                tempList.append(x)
        if(len(tempList) >= len(self.hashList)/3*2):
                self.hashList = [None] * len(self.hashList) * 2
                self.listSize = len(self.hashList)                
                #copying objects from temporary list to original hash list
                for item in tempList:
                    self.insertDouble(item)   
                
    # insert function
    def insertDouble(self, obj):        
        # calling hash function
        key = self.hash(obj[0]) 
        # step size calculation with random prime number
        step = 71 - (key % 71)
        # checking for key availability and adding the new object
        while(key < len(self.hashList)):            
            if self.hashList[key] == None:
                self.hashList[key] = obj
                #calling linearExceeding function
                self.doubleExceeding()
                return self.hashList
            else:
                # increment key value by step size
                key = key + step
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList)
        else:
            return False        
    
    def searchDouble(self, nic):
        # calling hash function
        key = self.hash(nic)        
        # step size calculation with random prime number
        step = 71 - (key % 71)
        # checking for key availability and presenting result set or returning False if not available in the list
        while(key < len(self.hashList)):
            if self.hashList[key] == None:
                return False            
            if(nic == self.hashList[key][0]):
                return [self.hashList[key][1], self.hashList[key][2], self.hashList[key][3]] 
            else:
                # increment key value by step size
                key = key + step
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList) 
        else:
            return False    
    
    def deleteDouble(self, nic):
        # calling hash function
        key = self.hash(nic)        
        # step size calculation with random prime number
        step = 71 - (key % 71)
        # checking for key availability and deletion or returning False if not found
        while(key < len(self.hashList)):
            if self.hashList[key] == None:
                return False            
            if(nic == self.hashList[key][0]):
                # making the index to None
                self.hashList[key] = None
                print("Deleted")
                return self.hashList
            else:
                # increment key value by step size
                key = key + step
                # when key is out of list indices
                if(key > len(self.hashList)-1):
                    key = key - len(self.hashList) 
        else:
            return False      
    
    # end of Double Hashing
    ####################################################################################################################        
```
